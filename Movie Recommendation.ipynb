```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

Merging the Two Datasets

```python
movies = movies.merge(credits, on='title')
```

We're only going to keep the relevant columns. We'll keep the genre, KW, ID, Title, Overview, Cast, Crew

```python
movies = movies[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

We need to get this into a better format now - Do not like the format.

```python
def dataInBetterShape(list_):
    data = []
    for i in ast.literal_eval(list_):
        data.append(i['name'])
    return data
```

```python
movies.iloc[0].keywords
```

```python
movies['genres'] = movies['genres'].apply(dataInBetterShape)
movies['keywords']=movies['keywords'].apply(dataInBetterShape)
#Doing this to just get the names
```

```python
def getCast(list_):
    data = []
    counter = 0
    for i in ast.literal_eval(list_):
        if counter!=3:
            data.append(i['name'])
            counter+=1
        else:
            break
    return data
    
```

```python
movies['cast']=movies['cast'].apply(getCast)
```

```python
movies.iloc[0].crew
```

We just need to extract the Directors here.

```python
def getDir(list_):
    data = []
    for i in ast.literal_eval(list_):
        if i['job'] == 'Director':
            data.append(i['name'])
            break
    return data
```

```python
movies['crew'] = movies['crew'].apply(getDir)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
def Convert(string):
    li = list(string.split(" "))
    return li
```

```python
movies['overview'] = movies['overview'].apply(Convert)
```

```python
movies.head()
```

Removing the spaces since that makes the Actors/Directors have seperate entities for First and Last names (idk if that makes sense)

```python
def remove_spaces_from_list(lst):
    return [string.replace(' ', '') for string in lst]
```

```python
movies['cast'] = movies['cast'].apply(remove_spaces_from_list)
```

```python
movies['genres'] = movies['genres'].apply(remove_spaces_from_list)
movies['crew'] = movies['crew'].apply(remove_spaces_from_list)
movies['keywords'] = movies['keywords'].apply(remove_spaces_from_list)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
new_df = movies[['id', 'title', 'tags']]
```

```python
def list_to_string_join(lst, separator=' '):
    return separator.join(str(item) for item in lst)
```

```python
new_df['tags'] = new_df['tags'].apply(list_to_string_join)
```

```python
new_df
```

```python
def convert_to_lowercase(input_string):
    return input_string.lower()
```

```python
new_df['tags'] = new_df['tags'].apply(convert_to_lowercase)
```

```python
new_df
```

Now we need to classify movies that are similar. Text Vectorization! Bag of words because it is 3AM and this started out as an intrusive thought and now I won't sleep until I finish it.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
movie_vec = cv.fit_transform(new_df['tags']).toarray()
```

```python
movie_vec
```

```python
cv.get_feature_names_out()
```

Now we apply stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
np.set_printoptions(threshold=np.inf)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(movie_vec).shape
```

```python
distance = cosine_similarity(movie_vec)
```

```python
distance[0]
```

```python
new_df
```

```python
sorted(list(enumerate(distance[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def Recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    similarity = distance[movie_index]
    movies_list = sorted(list(enumerate(similarity)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
Recommend("Batman Begins")
```
